Actividad de analitica de datos 

Operador de desempacado permite se define con un * este peremite que la funcion de python reciba multiples argumentos, almacenandolos en una tupla 

```python
import numpy as np  # Se importa la libreria numpy
import matplotlib.pyplot as plt

# Función de graficado con un número arbitrario de polinomios
def polinomios(*polinomios):
    x = np.linspace(-10, 10, 400)  # Generamos un rango de valores para x
    
    # Iteramos sobre cada polinomio en la lista
    for a, n in polinomios:
        y = a * x**n  # Calculo de  valores de y para el polinomio ax^n
        plt.plot(x, y, label=f'{a}x^{n}')  # Graficamos el polinomio con una etiqueta

    # Configuramos la gráfica
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.title("Gráfico de las funciones")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# se llama la función con múltiples polinomios
polinomios((1, 2), (2, 3), (-1, 4), (0.5, 10))
```
